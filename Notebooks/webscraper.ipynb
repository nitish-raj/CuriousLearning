```python
import requests
from bs4 import BeautifulSoup
```

```python

# URL of the HubSpot page you want to scrape
url = "https://offers.hubspot.com/business-idea-database?hubs_signup-url=offers.hubspot.com%2Fthank-you%2Fmfm-business-idea&hubs_signup-cta=typ-download"

# Send a GET request to the URL
response = requests.get(url)
```

```python
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find and extract the desired data from the HTML
# For example, to extract all the headings
headings = soup.find_all(["h3"])
for heading in headings:
    print(heading.text)

# Extract value in p class
descriptions = soup.find_all("p", class_="csol-full-width-cta-card-description")
for description in descriptions:
    print(description.text)

# Extract YouTube link
youtube_links = soup.find_all("a", class_="cl-button -primary -medium -light")
for link in youtube_links:
    print(link.get("href"))
```
